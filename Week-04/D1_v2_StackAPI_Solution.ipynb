# StackAPI

#### Import the necessary libraries here:

```python
# your code here
import pandas as pd
import numpy as np
from stackapi import StackAPI
from datetime import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
SITE = StackAPI('stackoverflow')
# SITE.max_pages=10
questions = SITE.fetch('questions', fromdate=datetime(2020,2,1), todate=datetime(2020,2,29))
questions
```

```python
answers = SITE.fetch('answers', fromdate=datetime(2020,2,1), todate=datetime(2020,2,29))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
from datetime import date

py_questions = SITE.fetch('questions', 
                       tagged= 'python', 
                       date=date.today(),
                       min = 5,
                       sort = 'votes')
py_questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
id_answers = SITE.fetch('answers/{ids}', ids=[6784, 6473])
id_answers
```
